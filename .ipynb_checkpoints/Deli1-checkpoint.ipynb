```python
from IPython.display import IFrame
ATPlink="https://www.tennisexplorer.com/ranking/atp-men/"
IFrame(ATPlink, width=900, height=500)
```

```python
import pandas as pd
ATPtables=pd.read_html(ATPlink, header=0, flavor='bs4')
                        
```

```python
ATPtables[1]
```

```python
DFatp=ATPtables[1].copy()
```

```python
DFatp.columns
```

```python
bye=[1]
DFatp.drop(columns=DFatp.columns[bye],inplace=True)
DFatp
```

```python
DFatp.columns.to_list()
```

```python
DFatp.columns.str.replace(pat="Name",
                           repl="Name(Male)", 
                           regex=True) 
```

```python
DFatp.columns.str.replace("Name","Name(Male)",regex=True).to_list()
```

```python
DFatp.columns=DFatp.columns.str.replace("Name","Name(Male)",regex=True).to_list()
```

```python
DFatp.Country.str.strip()
```

```python
DFatp.Country.value_counts()
```

```python
tableCounts=DFatp.Country.value_counts()
tableCounts
```

```python
tableCounts[tableCounts==10]
```

```python
MostPlayers=tableCounts[tableCounts==10].index.to_list()
MostPlayers
```

```python
DFatp[DFatp.Country.isin(MostPlayers)]
```
